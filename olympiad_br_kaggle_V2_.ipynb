## 0.1 Functions

```python
#pd.options.display.max_columns = 100
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)


def apply_label_encoder_test(df, label_encoder_path, column_name):
    # Carrega o codificador LabelEncoder treinado com os dados de treinamento
    with open(label_encoder_path, 'rb') as handle:
        le = pickle.load(handle)
        
    # Aplica o LabelEncoder nos dados de teste
    try:
        df[column_name + '_n'] = le.transform(df[column_name])
    except ValueError as e:
        # Encontra as categorias únicas nos dados de teste que não estão presentes nos dados de treinamento
        unseen_categories = set(df[column_name]) - set(le.classes_)

        # Adiciona as categorias únicas aos dados de treinamento
        le.classes_ = np.concatenate((le.classes_, list(unseen_categories)))

        # Aplica o LabelEncoder novamente nos dados de teste
        df[column_name + '_n'] = le.transform(df[column_name])
        
    # Remove a coluna original
    df = df.drop(column_name, axis=1)
    
    return df


def create_apply_min_max_scaler(df, column_name):

    # Cria uma cópia do DataFrame original
    df_scaled = df.copy()
    
    # Seleciona a coluna a ser escalonada
    column = df_scaled[column_name]
    
    # Cria uma instância do MinMaxScaler
    scaler = MinMaxScaler()
    
    # Aplica o escalonamento na coluna selecionada
    column_scaled = scaler.fit_transform(column.values.reshape(-1, 1))
    pickle.dump( scaler, open( 'data/pkl/mms_{}.pkl'.format(column_name), 'wb') )
    
    # Substitui a coluna original pela coluna escalonada no DataFrame copiado
    df_scaled[column_name] = column_scaled
    
    # Retorna o DataFrame com a coluna escalonada
    return df_scaled

def apply_min_max_scaler(df, column_name):
    scaler_path = 'data/pkl/mms_' + column_name + '.pkl'
    # Cria uma cópia do DataFrame original
    df_scaled = df.copy()
    
    # Seleciona a coluna a ser escalonada
    column = df_scaled[column_name]
    
    # Carrega o modelo de MinMaxScaler treinado anteriormente
    scaler = pickle.load(open(scaler_path, 'rb'))
    
    # Aplica o escalonamento na coluna selecionada
    column_scaled = scaler.transform(column.values.reshape(-1, 1))
    
    # Substitui a coluna original pela coluna escalonada no DataFrame copiado
    df_scaled[column_name] = column_scaled
    
    # Retorna o DataFrame com a coluna escalonada
    return df_scaled



def create_apply_robust_scaler(df, column_name):

    # Cria uma cópia do DataFrame original
    df_scaled = df.copy()
    
    # Seleciona a coluna a ser escalonada
    column = df_scaled[column_name]
    
    # Cria uma instância do RobustScaler
    scaler = RobustScaler()
    
    # Aplica o escalonamento na coluna selecionada
    column_scaled = scaler.fit_transform(column.values.reshape(-1, 1))
    pickle.dump( scaler, open( 'data/pkl/rbs_{}.pkl'.format(column_name), 'wb') )

    
    # Substitui a coluna original pela coluna escalonada no DataFrame copiado
    df_scaled[column_name] = column_scaled
    
    # Retorna o DataFrame com a coluna escalonada
    return df_scaled


def save_classification_metrics(y_test, y_pred, model_name, ciclo=1):
    """
    Salva as métricas do relatório de classificação em um arquivo CSV.

    Parâmetros
    ----------
    y_test : array-like of shape (n_samples,)
        Os rótulos verdadeiros do conjunto de teste.
    y_pred : array-like of shape (n_samples,)
        As predições do modelo para o conjunto de teste.
    ciclo : int, opcional (padrão=1)
        O número do ciclo atual. Se não informado, assume o valor 1.

    """
    if os.path.exists(model_name + '_relatorio.csv'):
        df = pd.read_csv(model_name + '_relatorio.csv')
        ciclo = df['ciclo'].max() + 1
    else:
        df = pd.DataFrame(columns=['categoria', 'precision', 'recall', 'f1-score', 'support', 'ciclo'])

    report = classification_report(y_test, y_pred, output_dict=True)
    df_novo = pd.DataFrame(report).transpose()[['precision', 'recall', 'f1-score', 'support']]
    df_novo['categoria'] = df_novo.index
    df_novo['ciclo'] = ciclo

    df = pd.concat([df, df_novo], axis=0)
    df.to_csv(model_name + '_relatorio.csv', index=False)
```

```python
def save_model_results(y_test, y_pred, csv_file_path):
    # Check if file exists
    try:
        df = pd.read_csv(csv_file_path)
        last_cycle = df['ciclo'].max()
        cycle = last_cycle + 1
    except FileNotFoundError:
        cycle = 1
    
    # Compute classification report
    report = classification_report(y_test, y_pred, output_dict=True)
    df_results = pd.DataFrame(report).transpose()
    
    # Add cycle column
    df_results['ciclo'] = cycle
    
    # Append to existing file or create new file
    try:
        df_results.to_csv(csv_file_path, mode='a', header=False)
    except FileNotFoundError:
        df_results.to_csv(csv_file_path)
```

```python
# definindo o número do ciclo
ciclo = 1

# verificando se o arquivo já existe
if os.path.exists('relatorio.csv'):
    # se o arquivo existe, carrega o último ciclo salvo
    df = pd.read_csv('relatorio.csv')
    ciclo = df['ciclo'].max() + 1
else:
    # se o arquivo não existe, cria um DataFrame vazio
    df = pd.DataFrame(columns=['categoria', 'precision', 'recall', 'f1-score', 'support', 'ciclo'])

# obtendo as métricas do relatório de classificação
report = classification_report(y_test, y_pred, output_dict=True)

# criando um novo DataFrame com as métricas do ciclo atual
df_novo = pd.DataFrame(report).transpose()[['precision', 'recall', 'f1-score', 'support']]
df_novo['categoria'] = df_novo.index
df_novo['ciclo'] = ciclo

# concatenando o novo DataFrame com o anterior
df = pd.concat([df, df_novo], axis=0)

# salvando o DataFrame em um arquivo CSV
df.to_csv('relatorio.csv', index=False)
```

```python
def save_classification_metrics(y_test, y_pred, ciclo=1):
    """
    Salva as métricas do relatório de classificação em um arquivo CSV.

    Parâmetros
    ----------
    y_test : array-like of shape (n_samples,)
        Os rótulos verdadeiros do conjunto de teste.
    y_pred : array-like of shape (n_samples,)
        As predições do modelo para o conjunto de teste.
    ciclo : int, opcional (padrão=1)
        O número do ciclo atual. Se não informado, assume o valor 1.

    Retorna
    -------
    None
    """
    if os.path.exists('relatorio.csv'):
        df = pd.read_csv('relatorio.csv')
        ciclo = df['ciclo'].max() + 1
    else:
        df = pd.DataFrame(columns=['categoria', 'precision', 'recall', 'f1-score', 'support', 'ciclo'])

    report = classification_report(y_test, y_pred, output_dict=True)
    df_novo = pd.DataFrame(report).transpose()[['precision', 'recall', 'f1-score', 'support']]
    df_novo['categoria'] = df_novo.index
    df_novo['ciclo'] = ciclo

    df = pd.concat([df, df_novo], axis=0)
    df.to_csv('relatorio.csv', index=False)
```
